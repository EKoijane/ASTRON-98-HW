```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. How Many Plots? 

# 1a. Write a function that takes three arguments (x data, an amplitude con-
#stant A, and a constant w for determining wavelength) and returns a
#sinusoid. 

def sinusoid(x, A, w):
    return A * np.sin(w * x)
x = np.linspace(0, 2 * np.pi, 1000)
A = 1
w = 2
y = sinusoid(x, A, w) 

#1b. Create an array to be your x data from 0 to 2π of 1000 points.

plt.plot(x,y)
plt.title("Original Sin Wave")
plt.xlabel('Raw Data - x')
plt.ylabel('Amplitude - y')
plt.show()

#1c. Create an array of 5 different amplitude A values, and a separate array of 5 different w values.
#1d. Create a figure object to plot your graphs onto.
#1e. Loop through each pair of A and w values. For each iteration, calculate the array of y data that correspond to those values, using the same x data
#array for each iteration. Plot each array of y data onto your figure (Hint: you can plot at the same time that you iterate).

x = np.linspace(0, 2 * np.pi, 1000)
Multi_A = np.array([2, 4 ,6, 8, 10])
Multi_w = np.array([3, 6, 9, 12, 15])
plt.figure(figsize=(10, 6))
plt.plot(x, sinusoid(x, 1, 2), label="Original A=1, w=2", color='black', linewidth=2)
for A, w in zip(Multi_A, Multi_w):
    y = sinusoid(x, A, w)
    plt.plot(x,y)
plt.title("Conglomerate Sin Wave")
plt.show()

#See Below
```

```python
#2 Randomness

#2a. First you must generate two lists of random numbers, each 40 elements long with values ranging from 0 to 100.

Data1 = np.random.randint(0, 100, size=40)
List1 = pd.DataFrame(Data1, columns=['Random List 1'])

Data2 = np.random.randint(0, 100, size=40)
List2 = pd.DataFrame(Data2, columns=['Random List 2'])

#2b. Make a plot figure with a specified figsize.
#2c. Plot each list of numbers as a separate line on the figure. One of the lines should be orange with a linewidth of 10.
#The other should be red with a dashed linestyle.
#Hint: The x values for each point should correspond to the index of the list. Don’t
#forget your legend and labels!

plt.figure(figsize=(10, 5))
plt.plot(List1, color = 'orange', linewidth = '10', label = "List 1")
plt.plot(List2, color = 'red', linestyle = 'dashed', label = "List 2")
plt.xlabel('List Index')
plt.ylabel('List Value')
plt.title("Double Random List Plot")
plt.legend(loc='upper right')
plt.show()
```
